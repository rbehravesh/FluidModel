```python
import networkx as nx
import matplotlib as plt
import numpy as np
import random
```

```python
FileName = './topology.txt'
G = nx.read_edgelist(
    FileName, 
    create_using=nx.DiGraph,
    nodetype=int,
    data=(('weight',float),)
)

ingress_nodes = [0, 1, 2, 12, 22, 24]
for n in G.nodes():
    if n in ingress_nodes:
        G.nodes[n]['color'] = 'red'
        G.nodes[n]['WWW'] = 'dc_core' + str(n)
        G.nodes[n]['shareCap'] = 1200000
        G.nodes[n]['shareCost'] = 1
        G.nodes[n]['lat'] =random.uniform(40.2647663628264, 40.2674002969022)
        G.nodes[n]['lon']=random.uniform(-3.75721197590123, -3.75380601283906)
    else:
        G.nodes[n]['color'] = 'blue'
        G.nodes[n]['WWW'] = 'dc_edge' + str(n)
        G.nodes[n]['shareCap'] = 10000
        G.nodes[n]['shareCost'] = 50
        G.nodes[n]['lat'] =random.uniform(40.253602, 40.27668)
        G.nodes[n]['lon']=random.uniform(-3.77504, -3.737746)

nx.write_gml(G, './newTopology.gml')
```

```python
nx.draw(G, with_labels=True)
```
